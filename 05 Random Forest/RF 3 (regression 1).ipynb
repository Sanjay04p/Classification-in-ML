```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('rock_density_xray.csv')
```

```python
df.head()
```

```python
df.columns=['signal','density']
```

```python
sns.scatterplot(df,x='signal',y='density')
```

```python
X=df['signal']
y=df['density']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

# Comparing all regression models for the dataset


# 1.Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
 # as we have a single feature sklearn confuses it has a series or dataframe so we get error
X=df['signal'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

* by seeing metric calculation it ensures the performfance is pretty good but when we plot we get the real results

```python
signal_range=np.arange(0,100)
```

```python
signal_pred=model.predict(signal_range.reshape(-1,1))
```

```python
sns.scatterplot(df,x='signal',y='density')

plt.plot(signal_pred)
```

* it is just predicting a straight line of avg nearly 2.2 for any signals so it cannot be a good model

```python
def run_model(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    
    pred=model.predict(X_test)
    
    mae=mean_absolute_error(y_test,pred)
    rmse=np.sqrt(mean_squared_error(y_test,pred))
    print(f'MAE: {mae}')
    print(f'RMSE: {rmse}')
    
    #plot result
    signal_range=np.arange(0,100)
    signal_pred=model.predict(signal_range.reshape(-1,1))
    
    sns.scatterplot(df,x='signal',y='density',color='black')
    plt.plot(signal_pred)
```

# 2.Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.pipeline import make_pipeline
```

```python
pipe =make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
```

```python
run_model(pipe,X_train,y_train,X_test,y_test)
```

```python
pipe =make_pipeline(PolynomialFeatures(degree=6),LinearRegression())
run_model(pipe,X_train,y_train,X_test,y_test)
```

* here degree 6 fits well only when we know our data feature dosen't exceed 100 here it may not fit well with new data points that exceeds signal 100

# 3.KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k_values=[1,5,10,30]
for n in k_values:
    model=KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

```python
model=KNeighborsRegressor(n_neighbors=30)
run_model(model,X_train,y_train,X_test,y_test)
```

* as k value increases our data fits better 'k value of upto 30 fits better' at k=1 model overfits

# 4.Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

* 1 decision tree overfits the model

# 5.Support vector regressor

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
model=SVR()

param_grid={'C':[0.01,0.1,1,5,10,100,1000],
           'gamma':['auto','scale'],
           }
grid=GridSearchCV(model,param_grid)
```

```python
run_model(grid,X_train,y_train,X_test,y_test)
```

* SVR fits well with the given data

# 6.Random Forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=10) # n_estimators even it is bigger it doesn't matter in this case
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```

* as there is only 1 feature it behave like decision tree 

# 7.Gradient boosting regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model=GradientBoostingRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

* performing similarly like decision tree

# 8.Ada boost regressor

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
model=AdaBoostRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

* performing bit like decision tree
