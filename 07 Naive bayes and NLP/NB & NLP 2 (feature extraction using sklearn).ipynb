```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
text=['This is a line',
     'This is another line',
     'Completely different line']
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer #avoids manually building BOW model
```

```python
cv=CountVectorizer()
```

```python
cv.fit_transform(text) # now it will treat each line as seperate document
```

* it has stored frequency of words in sparse matrix

```python
sparse_matrix=cv.fit_transform(text) # bag of words model
```

```python
sparse_matrix.todense() # to see the results 
```

* while dealing with large datasets we should not call these kind of functions bcoz it takes lot of space

```python
cv.vocabulary_ # index of all words in the above matrix
```

# stop word parameter

```python
cv1=CountVectorizer(stop_words='english')
```

```python
sparse_matrix1=cv1.fit_transform(text) 
```

```python
sparse_matrix1.todense()
```

```python
cv1.vocabulary_
```

# tfidfTransformer: it takes count of word frequency(BOW) and transform it into vector

```python
tfidf=TfidfTransformer()
```

```python
sparse_matrix
```

```python
# passing BOW to transform TF-IDF (BOW ---> TFIDF)
results=tfidf.fit_transform(sparse_matrix)
```

```python
results
```

```python
results.todense()
```

# this all process can be done in one step using TFIDF vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv=TfidfVectorizer()
```

```python
tv_result=tv.fit_transform(text)
```

```python
tv_result
```

```python
tv_result.todense()
```

* CountVectorizer + TfidfTransformer = TfidfVectorizer
* CountVectorizer ---> BOW
* TfidfTransformer ---> TF-IDF
* TfidfVectorizer ---> BOW + TF-IDF in single step
