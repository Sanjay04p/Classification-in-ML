```python
X=pd.get_dummies(df.drop('species',axis=1),dtype='int',drop_first=True)
```

```python
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

# creating a model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

# only 3 penguins are misclassified

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```

# 'value' is storing the no. of each species present i.e no. of each classes

```python
# function to report the above tree sample
def vis_model(model):
    model_pred=model.predict(X_test)
    print(classification_report(y_test,model_pred))
    print('\n')
    plt.figure(figsize=(8,8),dpi=200)
    plot_tree(model,feature_names=list(X.columns),filled=True);
```

```python
vis_model(model)
```

```python
# exploring hyperparameters
# 1. minimizing max_depth:
pruned_tree_model=DecisionTreeClassifier(max_depth=3)
pruned_tree_model.fit(X_train,y_train)
```

```python
vis_model(pruned_tree_model)
```

```python
# 2.max_leaf
max_leaf_model=DecisionTreeClassifier(max_leaf_nodes=3)
max_leaf_model.fit(X_train,y_train)
```

```python
vis_model(max_leaf_model)
```

```python
# 3.criterion default: 'gini'
entropy_tree_model=DecisionTreeClassifier(criterion='entropy')
entropy_tree_model.fit(X_train,y_train)
```

```python
vis_model(entropy_tree_model)
```

```python
entropy_tree_pred=entropy_tree_model.predict(X_test)
confusion_matrix(y_test,entropy_tree_pred)
```
